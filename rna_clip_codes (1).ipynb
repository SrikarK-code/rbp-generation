## dataset

```python
import torch
from torch.utils.data import Dataset
import numpy as np
from torch.nn.utils.rnn import pad_sequence

class RNARBPDataset(Dataset):
    def __init__(self, rna_embeddings, rbp_embeddings):
        """
        Args:
            rna_embeddings (numpy.ndarray): RNA embeddings of shape (num_samples, seq_len, emb_dim)
            rbp_embeddings (numpy.ndarray): RBP embeddings of shape (num_samples, seq_len, emb_dim)
        """
        self.rna_embeddings = rna_embeddings
        self.rbp_embeddings = rbp_embeddings

    def __len__(self):
        return len(self.rna_embeddings)

    def __getitem__(self, idx):
        rna_emb = torch.from_numpy(self.rna_embeddings[idx]).float()
        rbp_emb = torch.from_numpy(self.rbp_embeddings[idx]).float()
        return rna_emb, rbp_emb

def collate_fn(batch):
    """
    Collate function to pad sequences in batch to the same length.
    Pads with NaN values which can be masked in attention layers.

    Args:
        batch: List of tuples (rna_emb, rbp_emb)

    Returns:
        Tuple of padded tensors (rna_padded, rbp_padded)
    """
    # Separate RNA and RBP embeddings
    rna_embs, rbp_embs = zip(*batch)

    # Pad RNA embeddings
    rna_padded = pad_sequence(rna_embs, batch_first=True, padding_value=float('nan'))

    # Pad RBP embeddings
    rbp_padded = pad_sequence(rbp_embs, batch_first=True, padding_value=float('nan'))

    return rna_padded, rbp_padded

# Create attention mask for padding
def create_padding_mask(tensor):
    """
    Creates attention mask for padded sequences.

    Args:
        tensor: Input tensor with possible NaN padding

    Returns:
        Boolean mask where True indicates valid values and False indicates padding
    """
    return ~torch.isnan(tensor).any(dim=-1)  # Shape: (batch_size, seq_len)
```

## clip

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import random
from tqdm import tqdm
import wandb
import matplotlib.pyplot as plt
```

```python
class RNARBPCLIPProjectionHead(nn.Module):

   def __init__(self,input_dim,output_dim):
       super().__init__()
       self.skip=nn.Linear(input_dim,output_dim)
       self.layer_scale=nn.Parameter(torch.ones(1)*1e-4)
       self.projection=nn.Sequential(nn.Linear(input_dim,input_dim*2),nn.LayerNorm(input_dim*2),nn.GELU(),nn.Dropout(0.1),nn.Linear(input_dim*2,input_dim*2),nn.LayerNorm(input_dim*2),nn.GELU(),nn.Dropout(0.1),nn.Linear(input_dim*2,output_dim),nn.LayerNorm(output_dim))

   def forward(self,x):return self.skip(x)+self.layer_scale*self.projection(x)

class RNARBPCLIPEncoder(nn.Module):

   def __init__(self,embed_dim,num_layers=3):
       super().__init__()
       self.layers=nn.ModuleList([nn.TransformerEncoderLayer(d_model=embed_dim,nhead=8,dim_feedforward=embed_dim*4,dropout=0.1) for _ in range(num_layers)])
       self.layernorm=nn.LayerNorm(embed_dim)

   def forward(self, x, src_key_padding_mask=None):
       # src_key_padding_mask: (batch_size, seq_len)
       # True indicates positions that should be masked (padding)
       for layer in self.layers:
           x = layer(x, src_key_padding_mask=src_key_padding_mask)
       return self.layernorm(x)

class RNARBPCLIPModel(nn.Module):

   def __init__(self,rna_dim=120,rbp_dim=1280,projection_dim=512):
       super().__init__()
       self.rna_encoder=RNARBPCLIPEncoder(rna_dim)
       self.rbp_encoder=RNARBPCLIPEncoder(rbp_dim)
       self.rna_projection=RNARBPCLIPProjectionHead(rna_dim,projection_dim)
       self.rbp_projection=RNARBPCLIPProjectionHead(rbp_dim,projection_dim)
       self.logit_scale=nn.Parameter(torch.ones([])*np.log(1/0.07))

   def forward(self,rna_emb,rbp_emb):
       rna_mask=create_padding_mask(rna_emb).transpose(0, 1)
       rbp_mask=create_padding_mask(rbp_emb).transpose(0, 1)

       # Replace NaN values with zeros before processing
       rna_emb = torch.nan_to_num(rna_emb, 0.0)
       rbp_emb = torch.nan_to_num(rbp_emb, 0.0)

       # Pass masks to transformer layers
       rna_enc = self.rna_encoder(rna_emb, src_key_padding_mask=~rna_mask)
       rbp_enc = self.rbp_encoder(rbp_emb, src_key_padding_mask=~rbp_mask)

       # Rest of your forward pass remains the same
       rna_embed = F.normalize(self.rna_projection(rna_enc[:,0]), dim=-1)
       rbp_embed = F.normalize(self.rbp_projection(rbp_enc[:,0]), dim=-1)
       logit_scale = self.logit_scale.exp()
       similarity = torch.matmul(rna_embed, rbp_embed.t()) * logit_scale
       labels = torch.arange(similarity.size(0), device=similarity.device)
       loss = (F.cross_entropy(similarity, labels) + F.cross_entropy(similarity.t(), labels)) / 2
       return rna_embed, rbp_embed, loss

```

```python
def main():

    print("\nInitializing wandb...")
    wandb.init(
        project="rna-rbp-clip",
        config={
            "learning_rate": 1e-4,
            "batch_size": 32,
            "weight_decay": 0.01,
            "patience": 5,
            "max_epochs": 20
        }
    )
    print("Wandb initialized successfully")

    print("Initializing datasets...")
    train_dataset = RNARBPDataset(train_rna_embeddings, train_rbp_embeddings)
    val_dataset = RNARBPDataset(val_rna_embeddings, val_rbp_embeddings)
    print(f"Dataset sizes - Train: {len(train_dataset)}, Val: {len(val_dataset)}")

    print("\nCreating data loaders...")
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2, collate_fn=collate_fn)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2, collate_fn=collate_fn)
    print(f"Number of batches - Train: {len(train_loader)}, Val: {len(val_loader)}")

    print("\nSetting up device...")
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    print("\nInitializing model...")
    model = RNARBPCLIPModel().to(device)
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")

    class EarlyStopping:
        def __init__(self, patience=5, min_delta=0):
            self.patience = patience
            self.min_delta = min_delta
            self.counter = 0
            self.best_loss = None
            self.early_stop = False
            print(f"Early stopping initialized with patience {patience}")

        def __call__(self, val_loss):
            if self.best_loss is None:
                self.best_loss = val_loss
                print(f"Early stopping - First loss: {val_loss:.4f}")
                return False
            elif val_loss > self.best_loss - self.min_delta:
                self.counter += 1
                print(f"Early stopping - Counter increased to {self.counter}")
                if self.counter >= self.patience:
                    self.early_stop = True
                    print("Early stopping criterion met!")
            else:
                print(f"Early stopping - Counter reset (previous: {self.counter})")
                self.best_loss = val_loss
                self.counter = 0
                return True
            return False

    print("\nInitializing early stopping...")
    early_stopping = EarlyStopping(patience=5)

    print("\nSetting up optimizer and scheduler...")
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=0.01)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=20)
    print(f"Initial learning rate: {scheduler.get_last_lr()[0]}")

    print("\nCreating model directory...")
    os.makedirs('clip_models', exist_ok=True)

    def evaluate_model(model, data_loader):
        model.eval()
        total_loss = 0
        batch_losses = []
        print("\nStarting evaluation...")
        with torch.no_grad():
            for batch_idx, (rna, rbp) in enumerate(data_loader):
                rna, rbp = rna.to(device), rbp.to(device)
                print(f"Eval batch {batch_idx+1}/{len(data_loader)} - "
                      f"RNA shape: {rna.shape}, RBP shape: {rbp.shape}")

                _, _, loss = model(rna, rbp)
                batch_losses.append(loss.item())
                total_loss += loss.item()
                print(f"Batch {batch_idx+1} loss: {loss.item():.4f}")

        avg_loss = total_loss / len(data_loader)
        print(f"Evaluation complete - Average loss: {avg_loss:.4f}")
        print(f"Min batch loss: {min(batch_losses):.4f}, Max batch loss: {max(batch_losses):.4f}")
        return avg_loss

    def train_epoch(model, train_loader, optimizer):
        model.train()
        total_loss = 0
        batch_losses = []
        print("\nStarting training epoch...")

        for batch_idx, (rna, rbp) in enumerate(train_loader):
            rna, rbp = rna.to(device), rbp.to(device)
            print(f"Train batch {batch_idx+1}/{len(train_loader)} - "
                  f"RNA shape: {rna.shape}, RBP shape: {rbp.shape}")

            optimizer.zero_grad()
            _, _, loss = model(rna, rbp)
            loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            batch_losses.append(loss.item())
            total_loss += loss.item()
            print(f"Batch {batch_idx+1} loss: {loss.item():.4f}")

            if (batch_idx + 1) % 10 == 0:
                print(f"Average loss after {batch_idx + 1} batches: {total_loss/(batch_idx+1):.4f}")

        avg_loss = total_loss / len(train_loader)
        print(f"Epoch complete - Average loss: {avg_loss:.4f}")
        print(f"Min batch loss: {min(batch_losses):.4f}, Max batch loss: {max(batch_losses):.4f}")
        return avg_loss

    def train_with_early_stopping():
        print("\nStarting training with early stopping...")
        best_model_state = None
        best_val_loss = float('inf')

        for epoch in range(20):
            print(f"\n{'='*50}")
            print(f"Epoch {epoch+1}/20")
            print(f"Current learning rate: {scheduler.get_last_lr()[0]:.6f}")

            train_loss = train_epoch(model, train_loader, optimizer)
            val_loss = evaluate_model(model, val_loader)
            scheduler.step()

            wandb.log({
                'epoch': epoch,
                'train_loss': train_loss,
                'val_loss': val_loss,
                'learning_rate': scheduler.get_last_lr()[0]
            })

            print(f"\nEpoch {epoch+1} Summary:")
            print(f"Training Loss: {train_loss:.4f}")
            print(f"Validation Loss: {val_loss:.4f}")

            if val_loss < best_val_loss:
                improvement = best_val_loss - val_loss
                best_val_loss = val_loss
                best_model_state = model.state_dict()
                print(f"New best model! (Improvement: {improvement:.4f})")
                torch.save(best_model_state, f'clip_models/best_model.pt')
                print("Model saved to clip_models/best_model.pt")

            if early_stopping(val_loss):
                if early_stopping.early_stop:
                    print(f"\nEarly stopping triggered after {epoch+1} epochs")
                    break

        if best_model_state is not None:
            print("\nTraining complete - Loading best model...")
            model.load_state_dict(best_model_state)
            print(f"Loaded best model with validation loss: {best_val_loss:.4f}")
        return model

    print("\nStarting training process...")
    final_model = train_with_early_stopping()

    print("\nSaving final model...")
    torch.save(final_model.state_dict(), 'clip_models/final_model.pt')
    print("Final model saved to clip_models/final_model.pt")
    print("\nTraining complete!")

    print("\nClosing wandb...")
    wandb.finish()

if __name__ == "__main__":
    main()
```
